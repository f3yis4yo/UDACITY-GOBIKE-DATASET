```python
# load in the dataset into a pandas dataframe
# from the exploratory section i saved the cleaned dataframe to a new csv file and that is what I am going to load here 

df2 = pd.read_csv('bike_sharing_cleaned.csv')
df2.head()
```

## Age Distribution of Riders

Riders have ages that range from 18 to 141. However ages above 100 apper to be outliers, because that seem to be an impossible feat. Images below show the distribution of the age, and what it looks like after grouping the ages into different categories.
 

```python
binsize = 5
bins = range(10, df2.member_age.max() + 5, 5)

plt.figure(figsize = [12, 6])
plt.hist(data = df2, x = 'member_age', bins = bins)
plt.xlabel('age')
plt.ylabel('count')
plt.title('Age distribution by count')
plt.show()
```

```python
df2['age_group'] = pd.cut(df2['member_age'],[0, 30, 60, 90, 120, 150], 
                                labels=['very_young', 'young', 'old', 'very_old', 'Extremely_old'])

plt.figure(figsize=(10,8))
sb.countplot(data= df2, x = 'age_group', palette = 'gray')
plt.xlabel('age_group')
plt.ylabel('count')
plt.title('age_group by count');
```

##  Correlation between numeric values within the dataset

The numeric values of the dataste do not correlate with each other completely, however there is a negative correlation between age and birth year as expected. With time, there is no correlation with other variables in the dataset, which means the time spent during each rides is independent of any factor or numerical variable.

```python
data_stat = ['member_age', 'duration_min', "member_birth_year", "bike_id"]
g = sb.PairGrid(data = df2, vars = data_stat);
g.map(plt.scatter, alpha = 1/3);
```

## Correlation of all variables in the dataset using heatmap 

The correlation of the variables in the datset is shown in the heatmap below. It clearly confirms our observation above and makes it clear enough the independency of duration of bike rides in relation to other variables.


```python
plt.figure(figsize = (10,5), dpi = 200)
sb.heatmap(df2.corr(), annot = True, fmt = '.2f', cmap = 'viridis', center = 0)

plt.title(" What is the  Correlation between each variable in the dataset?")
plt.xticks(rotation = 90)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
